คำนวณว่า ช่วงเวลาต่อไปนี้ ห่างกันกี่ ชั่วโมง กี่นาที กี่วินาที (print ออกมาให้สวยงาม)

- 12:30:15 - 13:41:07
- 09:11:50 - 21:30:15
- 19:25:30 - 03:10:15

##ข้อ 1

````python
q1 = '12:30:15 - 13:41:07'
q2 = q1.split("-")
Q1 = q2[0].split(":")
Q2 = q2[1].split(":")
hours1 = int(Q1[0])*3600
minutes1 = int(Q1[1])*60
seconds1 = int(Q1[2])
hours2 = int(Q2[0])*3600
minutes2 = int(Q2[1])*60
seconds2 = int(Q2[2])
f1 =  hours1 + minutes1 + seconds1
f2 =  hours2 + minutes2 + seconds2
f3 = (f2-f1)
rh = int(f3/3600)
rm = int((f3%3600)/60)
rs = int(f3%60)
print(f' จาก {q1} ซึ่งห่างกันได้ {rh}:{rm}:{rs}')
````

##ข้อ 2

````python
q1 = '09:11:50 - 21:30:15'
q2 = q1.split("-")
Q1 = q2[0].split(":")
Q2 = q2[1].split(":")
hours1 = int(Q1[0])*3600
minutes1 = int(Q1[1])*60
seconds1 = int(Q1[2])
hours2 = int(Q2[0])*3600
minutes2 = int(Q2[1])*60
seconds2 = int(Q2[2])
f1 =  hours1 + minutes1 + seconds1
f2 =  hours2 + minutes2 + seconds2
f3 = (f2-f1)
rh = int(f3/3600)
rm = int((f3%3600)/60)
rs = int(f3%60)
print(f' จาก {q1} ซึ่งห่างกันได้ {rh}:{rm}:{rs}')
````

##ข้อ 3

````python
q1 = '19:25:30 - 03:10:15'
q2 = q1.split("-")
Q1 = q2[0].split(":")
Q2 = q2[1].split(":")
hours1 = int(Q1[0])*3600
minutes1 = int(Q1[1])*60
seconds1 = int(Q1[2])
hours2 = int(Q2[0])*3600
minutes2 = int(Q2[1])*60
seconds2 = int(Q2[2])
f1 =  hours1 + minutes1 + seconds1
f2 =  hours2 + minutes2 + seconds2
f3 = (f2+86400)-f1
rh = int(f3/3600)
rm = int((f3%3600)/60)
rs = int(f3%60)
print(f' จาก {q1} ซึ่งห่างกันได้ {rh}:{rm}:{rs}')
````
